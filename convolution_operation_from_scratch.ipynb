```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
original = cv2.resize(cv2.imread('img.jpg',0),(512,512))
```

```python
filtered = np.copy(original)
```

```python
size_x = filtered.shape[0]
size_y = filtered.shape[1]
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(original)
plt.show()
```

```python
filter = [[-10,0,10],[-20,0,20],[-10,0,10]]
weight = 1
```

```python
# iterating over the width of the image
for x in range(1, size_x-1):
  # iterating over the length of the image
  for y in range(1, size_y-1):
    # fitting the filter over the image and calculating all 9 pixels
    convolution = 0.0
    convolution = convolution + (original[x-1, y-1]*filter[0][0])
    convolution = convolution + (original[x, y-1]*filter[0][1])
    convolution = convolution + (original[x+1, y-1]*filter[0][2])
    convolution = convolution + (original[x-1, y]*filter[1][0])
    convolution = convolution + (original[x, y]*filter[1][1])
    convolution = convolution + (original[x+1, y]*filter[1][2])
    convolution = convolution + (original[x-1, y+1]*filter[2][0])
    convolution = convolution + (original[x, y+1]*filter[2][1])
    convolution = convolution + (original[x+1, y+1]*filter[2][2])
    convolution = convolution*weight
    if(convolution<0):
      convolution=0
    if(convolution>255):
      convolution=255
    
    filtered[x,y] = convolution   
       
```

```python
# filter = [[-1,0,1],[-2,0,2],[-1,0,1]]
# weight = 1


size_y = filtered.shape[1]
plt.grid(False)
plt.gray()
plt.axis('off')
plt.imshow(filtered)
plt.show()
```
